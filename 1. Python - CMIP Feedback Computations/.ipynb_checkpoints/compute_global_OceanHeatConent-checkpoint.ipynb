<h1> Function that computes global average thetoa values from non-averaged data that is available in longrunmip </h1>
Since datasets do not adhere to strict guidelines on saving data, this has to be done seperately for each individual model

```python
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import xarray as xr
import intake
import os
import warnings
import json

import tikzplotlib # for saving figures to tikz files

warnings.simplefilter('ignore')

%matplotlib inline
plt.rcParams['figure.figsize'] = 12,9
%config InlineBackend.figure_format = 'retina'
```

<h1> CCSM3 </h1>

```python
## Loading and computing
name = 'CCSM3'
control = '1530'
experiment = '2120'

# control
thetao_data = xr.open_dataset('longrunmip_data/global/thetao_ann_' + name + '_control_' + control + '.nc', decode_times=False)
thetao_weight = xr.ones_like(thetao_data['z_t'])*thetao_data['TAREA']
thetao_ctrl = (thetao_data['thetao'] * thetao_weight).sum(dim = ['nlat', 'nlon', 'z_t']) / thetao_weight.sum(dim = ['nlat', 'nlon', 'z_t'])

# experiment
thetao_data = xr.open_dataset('longrunmip_data/global/thetao_ann_'+ name + '_abrupt4x_'  + experiment + '.nc', decode_times=False)
thetao_abr = (thetao_data['thetao'] * thetao_weight).sum(dim = ['nlat', 'nlon', 'z_t']) / thetao_weight.sum(dim = ['nlat', 'nlon', 'z_t'])

## Saving
directory = 'longrunmip_data/global/' + name
if not os.path.isdir(directory):
    os.mkdir(directory)
thetao_ctrl.to_netcdf(directory + '/thetao_' + name + '_control_' + control + '.nc')
thetao_abr.to_netcdf(directory + '/thetao_' + name + '_abrupt4x_' + experiment + '.nc')

## PLOTTING to check correctness
plt.plot(thetao_ctrl, 'b-', label = 'control')
plt.plot(thetao_abr, 'r-', label = 'experiment')
```

<h1> HadGEM2 </h1>

```python
## Loading and computing
name = 'HadGEM2'
control = '239'
experiment = '1295'

# control
thetao_data = xr.open_dataset('longrunmip_data/global/thetao_ann_' + name + '_control_' + control + '.nc', decode_times=False)
thetao_weight =  np.cos(np.deg2rad(thetao_data['latitude'])) * xr.ones_like(thetao_data['longitude']) * xr.ones_like(thetao_data['depth'])
thetao_ctrl = (thetao_data['thetao'] * thetao_weight).sum(dim = ['latitude', 'longitude', 'depth']) / thetao_weight.sum(dim = ['latitude', 'longitude', 'depth'])

# experiment
thetao_data = xr.open_dataset('longrunmip_data/global/thetao_ann_'+ name + '_abrupt4x_'  + experiment + '.nc', decode_times=False)

N = len(thetao_data['time'])
thetao_abr = np.empty(N)
thetao_abr[:] = np.nan

for i in range(0, N):
    thetao_i = thetao_data['thetao'].isel(time = i)
    thetao_mean_i =  (thetao_i * thetao_weight).sum(dim = ['latitude', 'longitude', 'depth']) / thetao_weight.sum(dim = ['latitude', 'longitude', 'depth'])
    thetao_abr[i] = thetao_mean_i
    
thetao_abr = xr.DataArray(thetao_abr)


## Saving
directory = 'longrunmip_data/global/' + name
if not os.path.isdir(directory):
    os.mkdir(directory)
thetao_ctrl.to_netcdf(directory + '/thetao_' + name + '_control_' + control + '.nc')
thetao_abr.to_netcdf(directory + '/thetao_' + name + '_abrupt4x_' + experiment + '.nc')

## PLOTTING to check correctness
plt.plot(thetao_ctrl, 'b-', label = 'control')
plt.plot(thetao_abr, 'r-', label = 'experiment')
```

<h1> MPIESM12 </h1>

```python
## Loading and computing
name = 'MPIESM12'
control = '1500'
experiment = '1000'

# control
thetao_data = xr.open_dataset('longrunmip_data/global/thetao_ann_' + name + '_control_' + control + '.nc', decode_times=False)
thetao_weight =  np.cos(np.deg2rad(thetao_data['lat'])) * xr.ones_like(thetao_data['lon']) * xr.ones_like(thetao_data['depth'])

N = len(thetao_data['time'])
thetao_ctrl = np.empty(N)
thetao_ctrl[:] = np.nan

for i in range(0, N):
    thetao_i = thetao_data['thetao'].isel(time = i)
    thetao_mean_i =  (thetao_i * thetao_weight).sum(dim = ['x', 'y', 'depth']) / thetao_weight.sum(dim =  ['x', 'y', 'depth'])
    thetao_ctrl[i] = thetao_mean_i
    
thetao_ctrl = xr.DataArray(thetao_ctrl)


# experiment
thetao_data = xr.open_dataset('longrunmip_data/global/thetao_ann_'+ name + '_abrupt4x_'  + experiment + '.nc', decode_times=False)

N = len(thetao_data['time'])
thetao_abr = np.empty(N)
thetao_abr[:] = np.nan

for i in range(0, N):
    thetao_i = thetao_data['thetao'].isel(time = i)
    thetao_mean_i =  (thetao_i * thetao_weight).sum(dim = ['x', 'y', 'depth']) / thetao_weight.sum(dim =  ['x', 'y', 'depth'])
    thetao_abr[i] = thetao_mean_i
    
thetao_abr = xr.DataArray(thetao_abr)

## Saving
directory = 'longrunmip_data/global/' + name
if not os.path.isdir(directory):
    os.mkdir(directory)
thetao_ctrl.to_netcdf(directory + '/thetao_' + name + '_control_' + control + '.nc')
thetao_abr.to_netcdf(directory + '/thetao_' + name + '_abrupt4x_' + experiment + '.nc')

## PLOTTING to check correctness
plt.plot(thetao_ctrl, 'b-', label = 'control')
plt.plot(thetao_abr, 'r-', label = 'experiment')
```
