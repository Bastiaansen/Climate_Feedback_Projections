<h1> make_mode_plots </h1>

This jupyter notebook can be used to make the spatial distribution plots of the found spatial modes and the estimated equilibrium distribution from the abrupt4xCO2 experiment.

It requires the computed eigenmodes from the MATLAB codes in folder '2. MATLAB - Fitting'.

NOTE: this jupyter notebook is hard-coded to work with 3 eigenmodes (M=3)! If using a different amount of modes, the computation of the estimated equilibrium distribution need to be changed to be the sum of all M spatial modes.

<h1> Packages </h1>

```python
from matplotlib import pyplot as plt
import numpy as np
#import pandas as pd
import xarray as xr

import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.util import add_cyclic_point
```

```python
# Change fonts to make them consistent with LaTeX fonts
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix';
matplotlib.rcParams['font.family'] = 'STIXGeneral';
```

<h1> Load the computed eigenmodes </h1>

```python
eigenmode_folder = "../2. MATLAB - Fitting/2. Spatial Observables Fit/Eigenmodes/"


dtas_modes = xr.open_dataset(eigenmode_folder + "dtas_modes.nc")
dIMB_modes = xr.open_dataset(eigenmode_folder + "dIMB_modes.nc")
Planck_modes = xr.open_dataset(eigenmode_folder + "Planck_modes.nc")
LR_modes = xr.open_dataset(eigenmode_folder + "LR_modes.nc")
qLW_modes = xr.open_dataset(eigenmode_folder + "q_LW_modes.nc")
qSW_modes = xr.open_dataset(eigenmode_folder + "q_SW_modes.nc")
ALB_modes = xr.open_dataset(eigenmode_folder + "alb_modes.nc")
cloud_modes = xr.open_dataset(eigenmode_folder + "cloud_modes.nc")
dIMB_initial_forcing = xr.open_dataset(eigenmode_folder + "dIMB_initial_forcing.nc")
```

<h1> Functions to plot the modes </h1>

```python
def plot_contour(axis, Y, title, c_levels):
    axis.add_feature(cfeature.COASTLINE, color = 'lightgray')
    data, plot_lon = add_cyclic_point(Y.values, Y['lon'])
    plot_data = xr.DataArray(data, coords = [Y['lat'], plot_lon], dims = ["lat", "lon"])
    handle = plot_data.plot.contourf(ax = axis, transform = ccrs.PlateCarree(), cmap = 'RdBu_r', levels = c_levels, extend = 'both', add_colorbar = False)
    #title = title + " ($\\tau = " + str(Y['tau'].values) + " $year)"
    #plt.title(title, fontsize = 20)
    return handle

def plot_eigenmode(Y, suptitle, labeltext, c_levels, c_levels_ticks, file_name):
    fig = plt.figure(figsize = (20,20))
    
    #fig.suptitle(suptitle, fontsize = 32)
        
    ax1 = plt.subplot(131, projection = ccrs.PlateCarree())
    h1 = plot_contour(ax1, Y['Eigenmodes'].isel(tau = 0), 'mode 1', c_levels)
    plt.title("mode 1 ($\\tau_1 = 4.5$ year)", fontsize = 16)
    
    ax2 = plt.subplot(132, projection = ccrs.PlateCarree())
    h2 = plot_contour(ax2, Y['Eigenmodes'].isel(tau = 1), 'mode 2', c_levels)
    plt.title("mode 2 ($\\tau_2 = 127$ year)", fontsize = 16)
    
    ax3 = plt.subplot(133, projection = ccrs.PlateCarree())
    h3 = plot_contour(ax3, Y['Eigenmodes'].isel(tau = 2), 'mode 3', c_levels)
    plt.title("mode 3 ($\\tau_3 = 889$ year)", fontsize = 16)
        
    fig.subplots_adjust(wspace = 0.01)
    
    pos3 = ax3.get_position()
    pos2 = ax2.get_position()
    pos1 = ax1.get_position()
    cbar_ax = fig.add_axes([(pos1.x1+pos1.x0)/2, pos2.y0-0.03, (pos3.x0+pos3.x1-pos1.x0-pos1.x1)/2, 0.01])
    cb = fig.colorbar(h1, cax = cbar_ax, orientation = 'horizontal')
    cb.set_label(labeltext, fontsize = 22)
    cb.set_ticks(c_levels_ticks)
    cb.ax.tick_params(labelsize = 8)
    
    for c in h1.collections:
        c.set_edgecolor("face")
    for c in h2.collections:
        c.set_edgecolor("face")
    for c in h3.collections:
        c.set_edgecolor("face")
    
    plt.savefig(file_name, format='pdf', bbox_inches = 'tight', pad_inches = 0)
```

```python
c_levels = np.linspace(-21, 21, 256)
c_levels_ticks = np.array([0, 5, 10, 15, 20])
plot_eigenmode(dtas_modes, 'temperature (tas)', 'temperature increase $\Delta T$ (°C)', c_levels, c_levels_ticks, 'tas_modes.pdf')
```

<h1> Function to plot the Equilibrium Configurations </h1>

```python
def plot_equilibrium(Z, title, labeltext, c_levels, c_levels_ticks, file_name):
    fig = plt.figure(figsize = (20,20))
    
    Z1 = Z['Eigenmodes'].isel(tau = 0)
    Z2 = Z['Eigenmodes'].isel(tau = 1)
    Z3 = Z['Eigenmodes'].isel(tau = 2)
    
    Y = Z1 + Z2 + Z3   
                
    ax1 = plt.subplot(311, projection = ccrs.PlateCarree())
    ax1.add_feature(cfeature.COASTLINE, color = 'lightgray')
    data, plot_lon = add_cyclic_point(Y.values, Y['lon'])
    plot_data = xr.DataArray(data, coords = [Y['lat'], plot_lon], dims = ["lat", "lon"])
    handle = plot_data.plot.contourf(ax = ax1, transform = ccrs.PlateCarree(), cmap = 'RdBu_r', levels = c_levels, extend = 'both', add_colorbar = False)
    plt.title(title, fontsize = 22)
        
    fig.subplots_adjust(wspace = -0.1)
    
    pos = ax1.get_position()
    cbar_ax = fig.add_axes([pos.x1+0.01, pos.y0, 0.01, (pos.y1-pos.y0)])
    cb = fig.colorbar(handle, cax = cbar_ax, orientation = 'vertical')
    cb.set_label(labeltext, fontsize = 18)
    cb.set_ticks(c_levels_ticks)
    cb.ax.tick_params(labelsize = 8)
    
    for c in handle.collections:
        c.set_edgecolor("face")
    
    plt.savefig(file_name, format='pdf', bbox_inches = 'tight', pad_inches = 0)
```

```python
c_levels = np.linspace(-32, 32, 256)
c_levels_ticks = np.array([0, 10, 20, 30])
plot_equilibrium(dtas_modes, 'temperature increase $\Delta T^*$', '$\Delta T^*$ (°C)', c_levels, c_levels_ticks, 'tas_eq.pdf')
```

```python
def plot_initial_IMB(Z0, title, labeltext, c_levels, c_levels_ticks, file_name):
    fig = plt.figure(figsize = (20,20))
    
    Y = Z0['Initial_Forcing']   
                
    ax1 = plt.subplot(311, projection = ccrs.PlateCarree())
    ax1.add_feature(cfeature.COASTLINE, color = 'lightgray')
    data, plot_lon = add_cyclic_point(Y.values, Y['lon'])
    plot_data = xr.DataArray(data, coords = [Y['lat'], plot_lon], dims = ["lat", "lon"])
    handle = plot_data.plot.contourf(ax = ax1, transform = ccrs.PlateCarree(), cmap = 'RdBu_r', levels = c_levels, extend = 'both', add_colorbar = False)
    plt.title(title, fontsize = 22)
        
    fig.subplots_adjust(wspace = -0.1)
    
    pos = ax1.get_position()
    cbar_ax = fig.add_axes([pos.x1+0.01, pos.y0, 0.01, (pos.y1-pos.y0)])
    cb = fig.colorbar(handle, cax = cbar_ax, orientation = 'vertical')
    cb.set_label(labeltext, fontsize = 18)
    cb.set_ticks(c_levels_ticks)
    cb.ax.tick_params(labelsize = 8)
    

    for c in handle.collections:
        c.set_edgecolor("face")
    
    plt.savefig(file_name, format='pdf', bbox_inches = 'tight', pad_inches = 0)
```

```python
c_levels = np.linspace(-30, 30, 256)
c_levels_ticks = np.array([-30, -20, -10, 0, 10, 20, 30])
plot_initial_IMB(dIMB_initial_forcing, 'Initial forcing $F$', '$F$ ($W/m^2$)', c_levels, c_levels_ticks, 'IMB_initial.pdf')
```

```python
def plot_equilibrium_IMB(Z0,Z, title, labeltext, c_levels, c_levels_ticks, file_name):
    fig = plt.figure(figsize = (20,20))
    
    Z1 = Z['Eigenmodes'].isel(tau = 0)
    Z2 = Z['Eigenmodes'].isel(tau = 1)
    Z3 = Z['Eigenmodes'].isel(tau = 2)
        
    Y = Z1 + Z2 + Z3 + Z0['Initial_Forcing']
                    
    ax1 = plt.subplot(311, projection = ccrs.PlateCarree())
    ax1.add_feature(cfeature.COASTLINE, color = 'lightgray')
    data, plot_lon = add_cyclic_point(Y.values, Y['lon'])
    plot_data = xr.DataArray(data, coords = [Y['lat'], plot_lon], dims = ["lat", "lon"])
    handle = plot_data.plot.contourf(ax = ax1, transform = ccrs.PlateCarree(), cmap = 'RdBu_r', levels = c_levels, extend = 'both', add_colorbar = False)
    plt.title(title, fontsize = 22)
        
    fig.subplots_adjust(wspace = -0.1)
    
    pos = ax1.get_position()
    cbar_ax = fig.add_axes([pos.x1+0.01, pos.y0, 0.01, (pos.y1-pos.y0)])
    cb = fig.colorbar(handle, cax = cbar_ax, orientation = 'vertical')
    cb.set_label(labeltext, fontsize = 18)
    cb.set_ticks(c_levels_ticks)
    cb.ax.tick_params(labelsize = 8)
 

    for c in handle.collections:
        c.set_edgecolor("face")
    
    plt.savefig(file_name, format='pdf', bbox_inches = 'tight', pad_inches = 0)
```

```python
c_levels = np.linspace(-60, 60, 256)
c_levels_ticks = np.array([-30, -15, 0, 15, 30, 45, 60])
plot_equilibrium_IMB(dIMB_initial_forcing, dIMB_modes, 'equilibrium net radiation imbalance $\Delta N^*$', '$\Delta N^*$ ($W/m^2$)', c_levels, c_levels_ticks, 'IMB_eq.pdf')
```
